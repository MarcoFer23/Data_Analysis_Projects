```python
"""
The dataset insurance.csv contains data about people's characteristics and habits,
as well as the area where they live, in relation to individual medical care costs
used as health insurance premiums.

Display the dataset dimensions, a preview, and observe the column names.
What are the average charges by region? Are there significant differences?
What about smoker? And sex?
What are the statistical descriptors of bmi?
What are the minimum, mean, and maximum charges for the different quartiles of bmi values?
"""

import pandas as pd
import numpy as np

insurance = pd.read_csv("File_Data_Repository/insurance.csv", sep=",")
insurance
```

```python
# Let's print the dataset
print("The dataset 'insurance.csv' contains the following data:\n")
insurance
```

```python
# Let's print the number of rows and columns (dataset dimensions)
print("The dataset has the following dimensions (number of rows and columns respectively):")
insurance.shape
```

```python
# Let's print a sample of 10 rows
print("Below is a random preview of the data contained in the dataset:")
insurance.sample(10)
```

```python
# Let's print the column names
print("The dataset is composed of the following columns (their names are listed below):")
insurance.columns
```

```python
# 1. Calculate the average charges by region
average_charges_region = insurance.groupby('region')['charges'].mean().round(2)

print("\nBelow are the average values of 'charges' by 'region':\n")
average_charges_region
```

```python
# 2. Calculate the average charges depending on smoker status
average_charges_smoker = insurance.groupby('smoker')['charges'].mean().round(2)

print("\nBelow are the average values of 'charges' based on smoker status:\n")
average_charges_smoker
```

```python
# 3. Calculate the average charges by sex
average_charges_sex = insurance.groupby('sex')['charges'].mean().round(2)

print("\nBelow are the average values of 'charges' by 'sex':\n")
average_charges_sex
```

```python
# 4. Let's print statistical descriptors per bmi
bmi_stats = insurance['bmi'].describe().round(1)

print("\nBelow are the statistical descriptors per 'bmi':\n")
bmi_stats
```

```python
# 5. Minimum, mean, and maximum charges by bmi quartiles
quartiles = np.percentile(insurance['bmi'], [25, 50, 75])

quartile_differences = insurance.groupby(
    pd.cut(
        insurance['bmi'],
        bins=[0, quartiles[0], quartiles[1], quartiles[2], float('inf')]
    )
)['charges'].agg(['min', 'mean', 'max']).round(2)

print(
    "\nBelow are the Minimum, Mean, and Maximum values of "
    "'charges' for the different quartiles of 'bmi':"
)
quartile_differences
```

```python
# Alternative method
bmi_quartiles = pd.qcut(insurance['bmi'], 4)

statistics_by_quartile = insurance.groupby(
    bmi_quartiles
)['charges'].agg(['min', 'mean', 'max']).round(2)

statistics_by_quartile
```
