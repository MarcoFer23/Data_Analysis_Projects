```python
# Let's read and analyse Asia GdP database downloaded from Kaggle
import matplotlib.pyplot as plt
import pandas as pd

income_data = pd.read_csv("File_Data_Repository/income.csv", sep=",")
income_data
```

```python
# Check the dataset dimensions
income_data.shape
```

```python
# Check if there are missing values
income_data.isna().sum()
```

```python
# Let's replace missing values
income_data.fillna("N.d.", inplace=True)
```

```python
# No more missing values 
income_data.isna().sum()
```

```python
# Let's preview a 10 rows sample of the dataset
income_data.sample(10).sort_index()
```

```python
# Let's display column names
income_data.columns
```

```python
# Let's display a summary of the statistical descriptors for this version
income_data.describe().round(2)
```

```python
# Let's compare high-incomes by education level
income_data.groupby("education")["income >50K"].sum().round(2).sort_values(ascending=False)
```

```python
# Let's compare high-incomes by sex
income_data.groupby("sex")["income >50K"].sum().round(2).sort_values(ascending=False)
```

```python
# Let's compare average weekly working hours using groupby aggregation (for incomes higher (1) and lower (0) than 50K)
income_data.groupby("income >50K")["hours-per-week"].mean().round(2).sort_values(ascending=False)
```

```python
# Let's perform exploratory analysis of demographic age distribution using histogram visualization.
income_data["age"].hist(bins=30)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Num. Individuals")

plt.show()
```

```python
# Let's generare a correlation matrix on numerical features to identify relationships and potential predictors of income level.
numeric_cols = income_data.select_dtypes(include="number")
numeric_cols.corr().round(2)
```

```python
# Calculate the average ages per work class
avg_age_work_class = income_data.groupby('workclass')['age'].mean().round(2)

print("\nBelow are the average ages per workclass:\n")
avg_age_work_class.sort_values()
```

```python
# Calculate the average capital gain per education type
max_gain = income_data.groupby('education')['capital-gain'].mean().round(2)

print("\nBelow are the average capital gains per education type:\n")
max_gain.sort_values()
```

```python
# Calculate the average ages per occupation type
avg_age_occupation = income_data.groupby('occupation')['age'].mean().round(2)

print("\nBelow are the average ages per occupation:\n")
avg_age_occupation.sort_values()
```

```python
# Let's count how many high incomes there are within the dataset ( 1 == True, 0 == False) 
income_data["income >50K"].value_counts()
```

```python
#Alternative method to count how many high incomes there are (.sum() count only True values by default)
income_data["income >50K"].sum()
```

```python
#Let's check how many high incomes there are based on the 'occupation' type
income_by_occ = (
    income_data.groupby("occupation")["income >50K"]
      .sum()
      .sort_values(ascending=False)
)

income_by_occ
```

```python
# Let's use matplotlib to plot data evidence
income_by_occ.plot(kind="bar", figsize=(12,6))

plt.title("Number of People with Income higher than 50K, by Occupation")
plt.ylabel("Count")
plt.xlabel("Occupation Type")
plt.xticks(rotation=45)

plt.show()
```

```python
# Aggregated occupation-level income statistics to identify professions with the highest proportion of individuals earning above 50K
occupation_stats = (
    income_data.groupby("occupation")["income >50K"]
      .agg(["count", "sum", "mean"])
      .rename(columns={
          "count": "total_people",
          "sum": "num_above_50K",
          "mean": "pct_above_50K"
      })
)

occupation_stats["pct_above_50K"] *= 100

occupation_stats.sort_values("pct_above_50K", ascending=False).round(2)
```

```python
# Analyze high-income distribution across occupations using pandas crosstab aggregation
income_crosstab = pd.crosstab(
    income_data["occupation"],
    income_data["income >50K"],
    margins=False
    )
income_crosstab.sort_values(by=1, ascending=False)
```

```python
# Aggregated and ranked high-income individuals (>50K) by occupation using groupby
(
    income_data.groupby("occupation")["income >50K"]
      .sum()
      .sort_values(ascending=False)
)
```
