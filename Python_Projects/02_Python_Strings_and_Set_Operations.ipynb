```python
""" 
Let's write a program that asks the user to enter:

                • the number of liters of gasoline in the tank
                • the fuel efficiency in km per liter
                • the price of gasoline per liter

                    and then displays the cost per 100 km.
"""

# Data input

user_name = str(input("What's your name?"))
available_fuel = float(input("Hi " + user_name + "! How many litres of petrol are in the tank??"))
fuel_efficiency = float(input("How many kilometers can you drive on 1 liter?"))
fuel_price = float(input("What is the price of gasoline per liter?"))

# Calculating the distance in kilometers a car can travel with the available fuel

covered_distance = available_fuel * fuel_efficiency

# Calculating fuel costs for a 100-kilometer trip 

hundred_km_fuel_cost = (100 / fuel_efficiency) * fuel_price

# Displaying the result to the user (using operators and the .format() method simultaneously):

answer1 = str("Thank you, " + user_name + "! The car can travel {} km with the available fuel.")
answer2 = str("Fuel cost per 100 km is €{}.")

print(answer1.format(covered_distance))
print(answer2.format(hundred_km_fuel_cost))

#Alternatively, we can reduce the code as shown below

full_answer = str("Thank you, " + user_name + "! The car can travel {0} km with the available fuel.\nFuel cost per 100 km is€{1}.")

print(full_answer.format(covered_distance,hundred_km_fuel_cost))
```

```python
"""
Check whether the number of characters in each of the following strings is between 5 and 8:
• Windows
• Excel
• PowerPoint
• Word
"""

check_words = ["Samsung", "Windows", "Excel", "Powerpoint", "Word"]

for x in check_words:
    if 5 <= len(x) <= 8:
        pos_answer ="{} has 5 to 8 characters."
        print(pos_answer.format(x))
    else:
        neg_answer = "{} doesn't have 5 to 8 characters."
        print(neg_answer.format(x))
```
