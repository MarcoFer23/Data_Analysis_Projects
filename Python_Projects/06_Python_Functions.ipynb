```python
"""
Write a function that, given a list of numbers, returns the largest one without using the built-in max() function.
"""

numbers = input("Enter a list of integers separated by spaces: ").split()

def find_max(nums):
    max_value = int(nums[0])
    for x in nums:
        x = int(x)
        if x > max_value:
            max_value = x
    return max_value

print("The maximum value is:", find_max(numbers))
```

```python
"""
Write a function that, given a list of numbers, returns both the minimum and maximum (without using min() and max()).
"""

numbers = input("Enter a list of integers separated by spaces: ").split()

def find_max(nums):
    max_value = int(nums[0])
    for x in nums:
        x = int(x)
        if x > max_value:
            max_value = x
    return max_value

def find_min(nums):
    min_value = find_max(nums)
    for x in nums:
        x = int(x)
        if x < min_value:
            min_value = x
    return min_value

print("The minimum value is:", find_min(numbers), "\nThe maximum value is:", find_max(numbers))
```

```python
"""
Write a function that returns the two largest numbers in a list.
"""

numbers = input("Enter a list of integers separated by spaces: ").split()

def two_largest(nums):
    first = float("-inf")
    second = float("-inf")

    for x in nums:
        x = int(x)
        if x > first:
            second = first
            first = x
        elif first > x > second:
            second = x

    return first, second

if len(numbers) < 2:
    print("The list must contain at least two numbers!")
else:
    print("The two largest numbers are:", two_largest(numbers))
```

```python
"""
Write a function that takes a list of numbers and a number K as input;
it returns the average of all numbers greater than or equal to K.
If none exist, print an appropriate message.
"""

numbers = input("Enter a list of integers separated by spaces: ").split()
K = int(input("Enter an integer K: "))

def average_geq_k(nums, k):
    filtered = [int(x) for x in nums if int(x) >= k]

    if len(filtered) > 0:
        return sum(filtered) / len(filtered)
    else:
        return "No numbers greater than or equal to K were found."

print(average_geq_k(numbers, K))
```

```python
"""
Write a function that prints a simple bar chart using asterisks.
Each number in the list corresponds to a row of asterisks.
"""

numbers = input("Enter a list of integers separated by spaces: ").split()

def print_asterisks(nums):
    for x in nums:
        print("*" * int(x))

print_asterisks(numbers)
```

```python
"""
Create a function that takes a text as input and returns a dictionary
counting how many times each character appears.
"""

text = """Lisetta goes for a walk in the countryside; she is happy to collect flowers growing along the roadsides."""

def count_characters(text):
    freq = {}
    for ch in text:
        if ch not in freq:
            freq[ch] = 1
        else:
            freq[ch] += 1
    return freq

print("Character frequency:", count_characters(text))
```

```python
"""
Same as above, but exclude punctuation and spaces.
"""

import string

text = """Lisetta goes for a walk in the countryside; she is happy to collect flowers growing along the roadsides."""

def count_letters_only(text):
    freq = {}
    for ch in text:
        if ch.isalpha():
            if ch not in freq:
                freq[ch] = 1
            else:
                freq[ch] += 1
    return freq

print("Letter frequency:", count_letters_only(text))
```

```python
"""
Create a function that counts characters without distinguishing between uppercase and lowercase.
"""

text = """Lisetta goes for a walk in the countryside; she is happy to collect flowers growing along the roadsides."""

def count_case_insensitive(text):
    freq = {}
    for ch in text.lower():
        if ch not in freq:
            freq[ch] = 1
        else:
            freq[ch] += 1
    return freq

print("Case-insensitive character frequency:", count_case_insensitive(text))
```
