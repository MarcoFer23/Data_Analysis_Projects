```python
# Manually create a list containing the numbers from 0 to 5, store it in a variable, and display it on the screen.

numbers = [0,1,2,3,4,5]
print("Here are the numbers in the list: ",str(numbers))

# alternative (using a range):

range_list = range(0,6)
print("Here are the numbers in the list: ",list(range_list))
```

```python
# Given the variable school_name = "Hogwarts", print only first letter, then the first three letters only. Finally, capitalise all letters and print.

school_name = "Hogwarts"
print(school_name[0])
print(school_name[0:3])
print(school_name.upper())
```

```python
# Given the variable: x = 10, increase it by 2 and then multiply it by 3. Here are three different solutions:

x = 10
x = x + 2
x = x * 3
print(int(x))

x = 10
x += 2
x *= 3
print(int(x))

x = 10
x = x.__add__(2)
x = x.__mul__(3)
print(int(x))
```
