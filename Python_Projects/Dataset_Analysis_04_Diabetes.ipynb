```python
# Let's explore a dataset about diabetes cases

import pandas as pd
import math as m

db = pd.read_csv("File_Data_Repository/diabetes.csv", sep=",")
diabetes = pd.DataFrame(db)
diabetes
```

```python
# Let's split dataset data by patient's age
filter_1 = diabetes['Age (years)'].between(0, 18)
filter_2 = diabetes['Age (years)'].between(19, 29)
filter_3 = diabetes['Age (years)'].between(30, 39)
filter_4 = diabetes['Age (years)'].between(40, 49)
filter_5 = diabetes['Age (years)'] >= 50
```

```python
# Let's select as subset all patients between 30 and 39 years old
subset = diabetes[filter_3]
subset
```

```python
# Let's calculate the average blood pressure for patients over 50 years old
patients_over_50 = filter_5

average_blood_pressure = diabetes[patients_over_50]['Diastolic blood pressure (mm Hg)'].mean()
average_blood_pressure
```

```python
# Alternative method using .loc[] and no separate variable to filter data
average_blood_pressure = diabetes.loc[
    diabetes['Age (years)'] >= 50,
    'Diastolic blood pressure (mm Hg)'
].mean()
average_blood_pressure
```

```python
# Let's calculate the average blood pressure per patient age
for x in range(0, diabetes['Age (years)'].max()+1):
    age_filter = diabetes['Age (years)'] == x
    average_blood_pressure = diabetes[age_filter]['Diastolic blood pressure (mm Hg)'].mean()

    if m.isnan(average_blood_pressure):
        pass
    else:
        print(
            "For a patient aged:",
            x,
            "the average blood pressure is equal to:",
            round(average_blood_pressure, 2)
        )
```

```python
# Alternative and more efficient method using .group by() 
buckets = diabetes.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean()
print("Average blood pressure per age:","\n",buckets.round(2))
```
