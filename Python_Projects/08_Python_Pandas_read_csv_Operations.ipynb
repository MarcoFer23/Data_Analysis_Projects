```python
# Using the same data as before, we dig deeper into data evidence

import pandas as pd

# Load the dataset
db = pd.read_csv("[local-file-path]", sep=",")
amazon_db = pd.DataFrame(db)

# Display the dataframe
amazon_db

# Search for reviews containing the word "app"
column = 'reviewText'
search_value = 'app'

filter_mask = amazon_db[column].str.contains(search_value, case=False)
result = amazon_db[filter_mask].sum()

result

# Search for reviews containing the word "game"
column = 'reviewText'
search_value = 'game'

filter_mask = amazon_db[column].str.contains(search_value, case=False)
result = amazon_db[filter_mask].sum()

result

# Count values in the "Positive" column
column = 'Positive'
value_counts = amazon_db[column].value_counts()

value_counts

# Positive reviews containing the word "recommend"
filter_recommend = amazon_db['reviewText'].str.contains('recommend', case=False)
filter_positive = amazon_db['Positive'] == True

result = amazon_db[filter_recommend & filter_positive].value_counts()

result
result.sample(10)

# Negative reviews
filter_positive = amazon_db['Positive'] == False

result = amazon_db[filter_positive].value_counts()

result
result.sample(10)

# Total reviews containing the word "kindle"
filter_all = amazon_db['reviewText'].str.contains('kindle', case=False)
overall_result = amazon_db[filter_all]

print(overall_result.shape[0])

# Positive reviews containing the word "kindle"
filter_mask = amazon_db['reviewText'].str.contains('kindle', case=False)
filter_positive = amazon_db['Positive'] == True

positive_result = amazon_db[filter_mask & filter_positive]

print(positive_result.shape[0])

# Calculate negative reviews
negative_result = overall_result.shape[0] - positive_result.shape[0]

print(negative_result)

# Final sentiment evaluation
if negative_result < positive_result.shape[0]:
    print("The reviews are mostly positive.")
else:
    print("The reviews are mostly negative.")
```

```python
"""
The dataset insurance.csv contains data about people's characteristics and habits,
as well as the area where they live, in relation to individual medical care costs
used as health insurance premiums.

Display the dataset dimensions, a preview, and observe the column names.
What are the average charges by region? Are there significant differences?
What about smoker? And sex?
What are the statistical descriptors of bmi?
What are the minimum, mean, and maximum charges for the different quartiles of bmi values?
"""

import pandas as pd
import numpy as np

insurance = pd.read_csv(
    "C:[file-path]",
    sep=","
)

print("The dataset 'insurance.csv' contains the following data:\n", insurance)

print(
    "\nThe dataset has the following dimensions "
    "(number of rows and columns respectively):\n",
    insurance.shape
)

print(
    "\nBelow is a random preview of the data contained in the dataset:\n",
    insurance.sample(10)
)

print(
    "\nThe dataset is composed of the following columns "
    "(their names are listed below):\n",
    insurance.columns
)

# 1. Average charges by region
average_charges_region = insurance.groupby('region')['charges'].mean()

print("\nBelow are the average values of 'charges' by 'region':\n")
print(average_charges_region)

# 2. Significant differences by smoker status
average_charges_smoker = insurance.groupby('smoker')['charges'].mean()

print("\nBelow are the average values of 'charges' by 'smoker':\n")
print(average_charges_smoker)

# 3. Significant differences by sex
average_charges_sex = insurance.groupby('sex')['charges'].mean()

print("\nBelow are the average values of 'charges' by 'sex':\n")
print(average_charges_sex)

# 4. Statistical descriptors of bmi
bmi_stats = insurance['bmi'].describe()

print("\nBelow are the statistical descriptors for 'bmi':\n")
print(bmi_stats)

# 5. Minimum, mean, and maximum charges by bmi quartiles
quartiles = np.percentile(insurance['bmi'], [25, 50, 75])

quartile_differences = insurance.groupby(
    pd.cut(
        insurance['bmi'],
        bins=[0, quartiles[0], quartiles[1], quartiles[2], float('inf')]
    )
)['charges'].agg(['min', 'mean', 'max'])

print(
    "\nBelow are the Minimum, Mean, and Maximum values of "
    "'charges' for the different quartiles of 'bmi':"
)
print(quartile_differences)

# Suggested alternative by Vita
bmi_quartiles = pd.qcut(insurance['bmi'], 4)

statistics_by_quartile = insurance.groupby(
    bmi_quartiles
)['charges'].agg(['min', 'mean', 'max'])

print(statistics_by_quartile)
```

```python
"""
The dataset pokemon.csv contains a Pokémon database, with data such as
name, attack types, attack/defense/etc. values, and whether or not they are
legendary.

• Check the dataset dimensions, preview it, and inspect the column names;
• Is it plausible that the first column should be an index?
• Compare it with the index automatically assigned by Pandas:
  do they match?
• If not, set the first column as the index.
"""

import pandas as pd

pokemon = pd.read_csv("[file-path]", sep=",")
pokemon

pokemon.shape
pokemon.sample(10)
pokemon.columns

# Compare the index with the '#' column
if pokemon['#'].iloc[-1] == pokemon.shape[0] - 1:
    print("The final values of the index column and the '#' column match.")
else:
    pokemon = pokemon.set_index('#')
    print("The final values of the index column and the '#' column do not match.\nThe '#' column has been set as the index.")

# Alternative comparison
if len(set(pokemon['Name'])) == pokemon.shape[0] - 1:
    print("The final values of the index column and the '#' column match.")
else:
    pokemon = pokemon.set_index('Name')
    print("The final values of the index column and the '#' column do not match.\nThe 'Name' column has been set as the index.")
```

```python
"""
Exploring the same dataset as before:
• Are there duplicates, i.e., rows referring to the same Pokémon?
• For each value in Type 1, how many Pokémon are there?
• Which are the Legendary Pokémon?
• And which are the Grass Type 1 Legendary Pokémon?
• And Legendary Pokémon of Type 1 Ice or Fire?
• Sort the dataset by the Name column
• Set Name as the index (do not execute because there are 2 duplicate names)
• How do we access the stats of all Pokémon whose name starts with D?
• Which Generation 1 Pokémon have Attack > 50 and HP < 60?
• Are there null values?
"""

import pandas as pd

pokemon = pd.read_csv("[file-path]",sep=",")

pokemon

# Are there duplicates, i.e., rows referring to the same Pokémon?
duplicates = pokemon.duplicated()
if duplicates.any():
    print("There are duplicates in the dataset.")
else:
    print("There are no duplicates in the dataset.")

# For each value in Type 1, how many Pokémon are there?
count_pokemon_type1 = pokemon["Type 1"].value_counts()
print("\nCount of Pokémon by 'Type 1':\n")
print(count_pokemon_type1)

# Which are the Legendary Pokémon?
pokemon_legg = pokemon[pokemon["Legendary"] == True]
print("\nBelow are the names of all Legendary Pokémon:\n")
print(pokemon_legg[["Name", "Type 1", "Legendary"]])

# And which are the Grass Type 1 Legendary Pokémon?
legendary_grass = pokemon_legg[pokemon_legg["Type 1"] == "Grass"]
print("\nBelow are all Grass-type Legendary Pokémon:\n")
print(legendary_grass[["Name", "Type 1", "Legendary"]])

# And Legendary Pokémon of Type 1 Ice or Fire? [Including Electric, extra]
legendary_ice_fire = pokemon_legg[
    (pokemon_legg["Type 1"] == "Ice") |
    (pokemon_legg["Type 1"] == "Fire") |
    (pokemon_legg["Type 1"] == "Electric")
]

print("\nBelow are all Ice or Fire Legendary Pokémon:\n")
print(legendary_ice_fire[["Name", "Type 1", "Legendary"]])

# Sort the dataset by the Name column
pokemon.sort_values(by="Name", inplace=True)

# Set Name as the index
pokemon.set_index("Name", inplace=True)

# How do we access the stats of all Pokémon whose name starts with D?
pokemon_starting_d = pokemon[pokemon.index.str.startswith("D")]
print("\nBelow are all Pokémon whose name starts with 'D':\n")
print(pokemon_starting_d)

# Which Generation 1 Pokémon have Attack > 50 and HP < 60?
gen1 = pokemon[pokemon["Generation"] == 1]
filtered_pokemon = gen1[(gen1["Attack"] > 50) & (gen1["HP"] < 60)]

print("\nBelow are Generation 1 Pokémon with Attack > 50 and HP < 60:\n")
print(filtered_pokemon)

# Are there null values?
null_data = pokemon.isnull().sum()
print("\nNull values per column in the dataset:\n")
print(null_data)
```
