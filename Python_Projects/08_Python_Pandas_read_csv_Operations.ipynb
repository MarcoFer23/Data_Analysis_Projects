```python
# Let's explore another dataset, dealing with diabetes cases

import pandas as pd
import math as m

db = pd.read_csv("File_Data_Repository/diabetes.csv", sep=",")
diabetes = pd.DataFrame(db)
diabetes
```

```python
# To get a sample of 5 rows from the dataset
diabetes.sample(5) 
```

```python
# Let's split dataset data by patient's age
filter_1 = diabetes['Age (years)'].between(0, 18)
filter_2 = diabetes['Age (years)'].between(19, 29)
filter_3 = diabetes['Age (years)'].between(30, 39)
filter_4 = diabetes['Age (years)'].between(40, 49)
filter_5 = diabetes['Age (years)'] >= 50
```

```python
# Let's select as subset all patients between 30 and 39 years old
subset = diabetes[filter_3]
subset
```

```python
# Let's calculate the average blood pressure for patients over 50 years old
patients_over_50 = filter_5

average_blood_pressure = diabetes[patients_over_50]['Diastolic blood pressure (mm Hg)'].mean()
average_blood_pressure
```

```python
# Alternative method using .loc[] and no separate variable to filter data
average_blood_pressure = diabetes.loc[
    diabetes['Age (years)'] >= 50,
    'Diastolic blood pressure (mm Hg)'
].mean()
average_blood_pressure
```

```python
# Let's calculate the average blood pressure per patient age
for x in range(0, diabetes['Age (years)'].max()+1):
    age_filter = diabetes['Age (years)'] == x
    average_blood_pressure = diabetes[age_filter]['Diastolic blood pressure (mm Hg)'].mean()

    if m.isnan(average_blood_pressure):
        pass
    else:
        print(
            "For a patient aged:",
            x,
            "the average blood pressure is equal to:",
            round(average_blood_pressure, 2)
        )
```

```python
# Alternative and more efficient method using .group by() 
buckets = diabetes.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean()
print("Average blood pressure per age:","\n",buckets.round(2))
```

```python
"""
The dataset insurance.csv contains data about people's characteristics and habits,
as well as the area where they live, in relation to individual medical care costs
used as health insurance premiums.

Display the dataset dimensions, a preview, and observe the column names.
What are the average charges by region? Are there significant differences?
What about smoker? And sex?
What are the statistical descriptors of bmi?
What are the minimum, mean, and maximum charges for the different quartiles of bmi values?
"""

import pandas as pd
import numpy as np

insurance = pd.read_csv("File_Data_Repository/insurance.csv", sep=",")
insurance
```

```python
# Let's print the dataset
print("The dataset 'insurance.csv' contains the following data:\n")
insurance
```

```python
# Let's print the number of rows and columns (dataset dimensions)
print("The dataset has the following dimensions (number of rows and columns respectively):")
insurance.shape
```

```python
# Let's print a sample of 10 rows
print("Below is a random preview of the data contained in the dataset:")
insurance.sample(10)
```

```python
# Let's print the column names
print("The dataset is composed of the following columns (their names are listed below):")
insurance.columns
```

```python
# 1. Calculate the average charges by region
average_charges_region = insurance.groupby('region')['charges'].mean()

print("\nBelow are the average values of 'charges' by 'region':\n")
average_charges_region
```

```python
# 2. Calculate the average charges depending on smoker status
average_charges_smoker = insurance.groupby('smoker')['charges'].mean()

print("\nBelow are the average values of 'charges' based on smoker status:\n")
average_charges_smoker
```

```python
# 3. Calculate the average charges by sex
average_charges_sex = insurance.groupby('sex')['charges'].mean()

print("\nBelow are the average values of 'charges' by 'sex':\n")
average_charges_sex
```

```python
# 4. Let's print statistical descriptors per bmi
bmi_stats = insurance['bmi'].describe()

print("\nBelow are the statistical descriptors per 'bmi':\n")
bmi_stats
```

```python
# 5. Minimum, mean, and maximum charges by bmi quartiles
quartiles = np.percentile(insurance['bmi'], [25, 50, 75])

quartile_differences = insurance.groupby(
    pd.cut(
        insurance['bmi'],
        bins=[0, quartiles[0], quartiles[1], quartiles[2], float('inf')]
    )
)['charges'].agg(['min', 'mean', 'max'])

print(
    "\nBelow are the Minimum, Mean, and Maximum values of "
    "'charges' for the different quartiles of 'bmi':"
)
quartile_differences
```

```python
# Alternative method
bmi_quartiles = pd.qcut(insurance['bmi'], 4)

statistics_by_quartile = insurance.groupby(
    bmi_quartiles
)['charges'].agg(['min', 'mean', 'max'])

statistics_by_quartile
```

```python
"""
The dataset pokemon.csv contains a Pokémon database, with data such as
name, attack types, attack/defense/etc. values, and whether or not they are
legendary.

• Check the dataset dimensions, preview it, and inspect the column names;
• Is it plausible that the first column should be an index?
• Compare it with the index automatically assigned by Pandas:
  do they match?
• If not, set the first column as the index.
"""

import pandas as pd

pokemon = pd.read_csv("File_Data_Repository/pokemon.csv", sep=",")

#Display the full dataset
pokemon.sort_index()
```

```python
# Check the dataset dimensions
pokemon.shape
```

```python
# Let's preview a 10 rows sample of the dataset
pokemon.sample(10).sort_index()
```

```python
# Let's display column names
pokemon.columns
```

```python
# Are there null values within the 'pokemon' dataset?
null_data = pokemon.isnull().sum()
print("\nNull values per column in the dataset:\n")
null_data
```

```python
# Are there duplicates, i.e., rows referring to the same Pokémon?
duplicates = pokemon.duplicated()
if duplicates.any():
    print("There are duplicate rows in the dataset.")
else:
    print("There are no duplicate rows in the dataset.")
```

```python
# For each value in Type 1, how many Pokémon are there?
count_pokemon_type1 = pokemon["Type 1"].value_counts()

print("\nCount of Pokémon by 'Type 1':\n")
count_pokemon_type1
```

```python
# Which are the Legendary Pokémon?
pokemon_legg = pokemon[pokemon["Legendary"] == True]

print("\nBelow are the names of all Legendary Pokémon:\n")
pokemon_legg[["Name", "Type 1", "Legendary", "Generation"]].sort_index(ascending=True)
```

```python
# Which are the Grass Type 1 Legendary Pokémon?
pokemon_legg = pokemon[pokemon["Legendary"] == True]
legendary_grass = pokemon_legg[pokemon_legg["Type 1"] == "Grass"]

print("\nBelow are all Grass-type Legendary Pokémon:\n")
legendary_grass[["Name", "Type 1", "Legendary", "Generation"]].sort_index(ascending=True)
```

```python
# Which Legendary Pokémon are Type 1 Psychic, Water, or Dragon ?
pokemon_legg = pokemon[pokemon["Legendary"] == True]
legendary_p_w_d = pokemon_legg[
    (pokemon_legg["Type 1"] == "Psychic") |
    (pokemon_legg["Type 1"] == "Water") |
    (pokemon_legg["Type 1"] == "Dragon")
]

print("\nBelow are Psychic, Water or Dragon Legendary Pokémon:\n")
legendary_p_w_d[["Name", "Type 1", "Legendary", "Generation"]].sort_index(ascending=True)
```

```python
# Which Legendary Pokémon are Type 1 Water, Ice, Fire or Electric ?
pokemon_legg = pokemon[pokemon["Legendary"] == True]
legendary_wa_ic_fi_el = pokemon_legg[
    (pokemon_legg["Type 1"] == "Water") |
    (pokemon_legg["Type 1"] == "Ice") |
    (pokemon_legg["Type 1"] == "Fire") |
    (pokemon_legg["Type 1"] == "Electric")
]

print("\nBelow are Water, Ice, Fire or Electric Legendary Pokémon:\n")
legendary_wa_ic_fi_el[["Name", "Type 1", "Legendary", "Generation"]].sort_index(ascending=True)
```

```python
# Which Legendary Pokémon are from the 2nd generation?
pokemon_legg = pokemon[pokemon["Legendary"] == True]
legendary_2nd_gen = pokemon_legg[pokemon_legg["Generation"] == 2]

print("\nBelow are Water, Ice, Fire or Electric Legendary Pokémon:\n")
legendary_2nd_gen[["Name", "Type 1", "Legendary", "Generation"]].sort_index(ascending=True)
```

```python
# Which Generation 1 Pokémon have Attack > 50 and HP < 60?
gen1 = pokemon[pokemon["Generation"] == 1]
filtered_pokemon = gen1[(gen1["Attack"] > 50) & (gen1["HP"] < 60)]

print("\nBelow are Generation 1 Pokémon with Attack > 50 and HP < 60:\n")
filtered_pokemon[["Name", "Type 1", "Generation"]].sort_index(ascending=True)
```

```python
# Access the stats of all Pokémon whose name starts with D?

pokemon_name_starting_d = pokemon[pokemon["Name"].str.startswith("D")]
print("\nBelow are Pokémon whose name starts with 'D':\n")
pokemon_name_starting_d
```

```python
# Compare the index with the '#' column
if pokemon['#'].iloc[-1] == pokemon.shape[0] - 1:
    print("The final values of the index column and the '#' column match.")
else:
    pokemon = pokemon.set_index('#')
    print("The final values of the index column and the '#' column do not match.\nThe '#' column has been set as the index.")
```
