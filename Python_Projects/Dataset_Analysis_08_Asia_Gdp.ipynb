```python
# Let's read and analyse Asia GdP database downloaded from Kaggle

import pandas as pd
income_data = pd.read_csv("File_Data_Repository/income.csv", sep=",")
income_data
```

```python
# Check the dataset dimensions
income_data.shape
```

```python
# Let's preview a 10 rows sample of the dataset
income_data.sample(10).sort_index()
```

```python
# Let's display column names
income_data.columns
```

```python
# Let's display a summary of the statistical descriptors for this version
income_data.describe().round(2)
```

```python
# 1. Calculate the average ages per work class
avg_age_work_class = income_data.groupby('workclass')['age'].mean().round(2)

print("\nBelow are the average ages per workclass:\n")
avg_age_work_class.sort_values()
```

```python
# 2. Calculate the average capital gain per education type
max_gain = income_data.groupby('education')['capital-gain'].mean().round(2)

print("\nBelow are the average capital gains per education type:\n")
max_gain.sort_values()
```

```python
#3. Calculate the average ages per occupation type
avg_age_occupation = income_data.groupby('occupation')['age'].mean().round(2)

print("\nBelow are the average ages per occupation:\n")
avg_age_occupation.sort_values()
```
