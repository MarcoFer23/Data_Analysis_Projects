```python
# Given the string: school_name = “Hogwarts”. Print each character of the string in uppercase, one per line, using a while loop.

school_name = "Hogwarts"
i = 0

while i < len(school_name):
    print(school_name[i].upper())
    i += 1
```

```python
# Print all numbers from 1 to 20 on the screen using a while loop.

i = 1
while i <= 20:
    print(i)
    i += 1

# Alternative (1):

i = 1
while i <= 20:
    print(i," ", end="\n")
    i += 1

# Alternative (2):
    
i = 1
risultato = ""

while i <= 20:
    risultato += str(i) + "\n"
    i += 1

print(risultato.strip("\n"))

# Alternative (3):

i = 1
risultato = ""

while i <= 20:
    risultato += str(i) + "\n"
    i += 1

print(risultato.strip())
```

```python
# Calculate and print the first 10 powers of 2 using a while loo.

i = 2
x = 0 

result = ""

while x < 10:
    result  += str(i**x) + "\n"
    x += 1

print(result.strip("\n"))

# Enriched version:

i = 2
x = 0 

result = ""

while x <10:
    result  += str(i) + "^" + str(x) + " = " + str(i**x) + "\n"
    x += 1

print(result.strip("\n"))
```

```python
# Calculate and print the first N powers of 2 using a while loop, prompting the user to enter N.

i = 2
x = 0 
N = int(input("Enter a value N to start the calculation."))
result = ""

while x < N:
    result += str(i**x) + "\n"
    x += 1

print(result.strip("\n"))

# Enriched version:

i = 2
x = 0 
N = int(input("Enter a value N to start the calculation."))

result = ""

while x < N:
    result  += str(i) + "^" + str(x) + " = " + str(i**x) + "\n"
    x += 1

print(result.strip("\n"))
```

```python
"""
Part 1:
Given two lists, one of students and one of courses:

students = [“Alex”, “Bob”, “Cindy”, “Dan”, “Emma”, “Faith”,
"Grace", “Henry”]
courses = [“Cybersecurity”, “Data Analyst”, “Backend”,
“Frontend”, “Data Analyst”, "Backend", “Frontend”]

Verify that both lists have the same length, and print to the screen
a message telling us whether this is the case or not.

Part 2:

Fill in the missing data in the course list, given that:

Emma is taking Data Analyst
Faith is taking Backend
Grace is taking Frontend
Henry is taking Cybersecurity

We will fill in the missing data one at a time using the method for appending
to the end of lists, then we will check that they are the same length, and if they are,
we will print the course list.
"""

# Part 1

students = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]
courses = ["Cybersecurity", "Data Analyst", "Backend", "Frontend", "Data Analyst", "Backend"]

if len(students) == len(courses):
    print("The two lists have the same length.")
else:
    print("The two lists don't have the same length.")

# Part 2

students = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]
courses = ["Cybersecurity", "Data Analyst", "Backend", "Frontend", "Data Analyst", "Backend"]

courses.append("Frontend")
courses.append("Cybersecurity")

if len(students) == len(courses):
    print("The two lists have the same length.")
else:
    print("The two lists don't have the same length.")
```

```python
"""
Write a program that prompts the user to enter a string and displays the first 3 characters,
followed by an ellipsis (…), and then the last 3 characters.
Be sure to handle all special cases, i.e., implement custom 
solutions for strings shorter than 6 characters.
"""

input_string_text = str(input("Do you ride motorcycles? If yes, enter your motorcycle's brand and model."))

if len(input_string_text) >= 6:
    print(input_string_text[0:3] + "..." + input_string_text[-3:])

else:
    print("Please use at least 6 characters.")
```

```python
# Store and print all factors of a given number.

input_num = int(input("Enter any integer N."))

list = []

div_by = 2

while input_num != 1:
    if input_num % div_by == 0:
        list.append(div_by)
        input_num /= div_by
    else:
        div_by += 1  

print(list)
```
