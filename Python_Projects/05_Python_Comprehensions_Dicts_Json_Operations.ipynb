```python
"""
We have a list of numbers:

numbers = [4, 10, 50, 100, 128, 71, 46]

Create a new list whose elements are the same as numbers, each increased by 10, using a list comprehension.
"""

numbers = [4, 10, 50, 100, 128, 71, 46]
plus_10_numbers = [n + 10 for n in numbers]

print(plus_10_numbers)
```

```python
"""
We have a list of numbers:

numbers = [4, 10, 50, 100, 128, 71, 46]

Create a new list whose elements are the same as numbers but doubled, using a list comprehension.
"""

numbers = [4, 10, 50, 100, 128, 71, 46]
doubled_numbers = [n * 2 for n in numbers]

print(doubled_numbers)
```

```python
"""
We have a list of names:

names = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]

Create a new list whose elements are the initials of the names, using a list comprehension.
"""

names = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]
initials_names = [name[0] for name in names]

print(initials_names)
```

```python
"""
We have a list of names:

names = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]

Create a new list whose elements are the initials of the names followed by a dot, using a list comprehension.
"""

names = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry"]
initials_names = [name[0] + "." for name in names]

print(initials_names)
```

```python
"""
We have a list of Italian tax codes (Codice Fiscale):

tax_ids = ["ABCDEF95G01A123B", "GHIJKL91M02A321C", "MNOPQR89S03A456D", "STUVWX95Z04A654E", "XYZABC01D05A789F", "DEFGHI95J06A987G"]

Create two lists:
- one containing (for each CF) only the characters related to the name
- one containing (for each CF) only the characters related to the surname

Both must be created using list comprehensions.
"""

tax_ids = ["ABCDEF95G01A123B", "GHIJKL91M02A321C", "MNOPQR89S03A456D", "STUVWX95Z04A654E", "XYZABC01D05A789F", "DEFGHI95J06A987G"]

# Extract characters corresponding to the "name" part (positions 3 to 5)
first_name_char = [n[3:6] for n in tax_ids]

# Extract characters corresponding to the "surname" part (positions 0 to 2)
last_name_char = [n[0:3] for n in tax_ids]

print(first_name_char, last_name_char)
```

```python
"""
We have a list of students:

students = ["Alex", "Bob", "Cindy", "Dan", "Emma",
             "Faith", "Grace", "Henry", "Isabelle", "John"]

We want to divide the students into two teams for a Uno tournament as follows:
the first half goes to one team, and the second half goes to the other team.

Create two lists for each team and display them at the end.
"""

students = ["Alex", "Bob", "Cindy", "Dan", "Emma",
            "Faith", "Grace", "Henry", "Isabelle", "John"]

squad_1 = [s for s in students[:len(students)//2]]
squad_2 = [s for s in students[len(students)//2:]]

print(squad_1, squad_2)
```

```python
"""
We have a list of price strings in dollars that were mistakenly written
with the euro symbol instead:

prices = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"]

Replace the euro symbol (€) with the dollar symbol ($) for each string
in the list using a list comprehension.
"""

prices = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"]

dollar_prices = [p.replace("€", "$") for p in prices]

print(dollar_prices)
```

```python
"""
We have a list of students:

students = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]

We want to divide the students into two teams for a Uno tournament as follows:
we select names in even positions for one team, and names in odd positions for the other team.

We create two lists for each team, and finally display them.
"""

students = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]

# Split using slicing (clean and efficient way)

team_1 = students[::2]   # even positions (0, 2, 4, ...)
team_2 = students[1::2]  # odd positions (1, 3, 5, ...)

print(team_1, team_2)
```

```python
"""
We have a list of earnings over the last 12 months (from January to December):

earnings = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

We need to compare, printing everything, each month's earnings with the average of previous months.

Example output:
Month 1: 100 €
Month 2: 90 € (previous avg: 100 €)
Month 3: 70 € (previous avg: 95 €)
"""

earnings = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

# First month has no previous average
print("Month 1:", str(earnings[0]) + " €")

# From second month onward, compute previous average
print(
    "".join([
        "Month " + str(pos + 1) + ": " +
        str(earnings[pos]) + " € (previous avg: " +
        str(sum(earnings[:pos]) // pos) + " €)\n"
        for pos in range(1, len(earnings))
    ])
)
```
