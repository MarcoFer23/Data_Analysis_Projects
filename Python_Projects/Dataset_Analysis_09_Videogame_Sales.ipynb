The dataset 'vgsales.csv', available on Kaggle, contains data about videogame sames in different markets, and allows us to explore data and compare sales between console platforms, videogame genre, publisher, etc. Let's explore available data and generate graphs in Plotly Express; please note: Publisher does not mean "Videogame Developer", so videogames like Final Fantasy X, developed by SquareSoft, might have been published by other companies in certain markets.

```python
# Let's read and analyse the videogame sales database downloaded from Kaggle
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

vg_sales = pd.read_csv("File_Data_Repository/vgsales.csv", sep=",")
vg_sales
```

```python
# Check the dataset dimensions
vg_sales.shape
```

```python
# Check if there are missing values
vg_sales.isna().sum()
```

```python
# For the time being, let's ignore null values within the Year column but let's fill "Publisher" null values 
vg_sales["Publisher"].fillna("Unknown", inplace=True)
```

```python
# No more missing values within the "Publisher" column
vg_sales.isna().sum()
```

```python
# Let's preview a 10 rows sample of the dataset
vg_sales.sample(10).sort_index()
```

```python
# Let's display column names
vg_sales.columns
```

```python
# Let's display a summary of the statistical descriptors for this version
vg_sales.describe().round(2)
```

```python
# Let's compare videogame global sales per Platform
vg_sales.groupby("Platform")["Global_Sales"].sum().round(2).sort_values(ascending=False)
```

```python
# Let's compare videogame global sales per Year
vg_sales.groupby("Year")["Global_Sales"].sum().round(2).sort_values(ascending=False)
```

```python
# Let's compare videogame global sales per Genre
vg_sales.groupby("Genre")["Global_Sales"].sum().round(2).sort_values(ascending=False)
```

```python
# Let's compare videogame global sales per Publisher
vg_sales.groupby("Publisher")["Global_Sales"].sum().round(2).sort_values(ascending=False)
```

We can use data visualization libraries to plot available data evidence; for the time being, let's focus on best selling videogames publishers, although we could use the same approach to plot graphs about global sales per platform, genre, etc.

```python
# Let's plot the 10 best selling videogame publishers (worldwide sales)
publisher_global_sales = vg_sales.groupby("Publisher")["Global_Sales"].sum().round(2).sort_values(ascending=False).reset_index()

fig = px.bar(
    publisher_global_sales.head(10),
    x="Global_Sales",
    y="Publisher",
    title="15 Best Videogame Publishers by Global Sales"
)

fig.update_layout(
    
    yaxis=dict(categoryorder="total ascending"),
    xaxis_title="Global Sales (Millions)",
    yaxis_title="Publisher",
)

fig.show()
```

```python
# Let's use a treemap plot to show Publishers market dominance by Global Sales
publisher_global_sales = vg_sales.groupby("Publisher")["Global_Sales"].sum().round(2).sort_values(ascending=False).reset_index()

fig = px.treemap(
    publisher_global_sales,
    path=["Publisher"],
    values="Global_Sales",
    title="Publisher Market Share by Global Sales",
    
    # Additional data available in hover
    hover_data={
        "Global_Sales": ":.2f"
    }
)

# Customise Text inside treemap rectangles
fig.update_traces(
    texttemplate="<b>%{label}</b><br>%{percentRoot:.1%}",
    textfont_size=16
)

# Customise hover popup
fig.update_traces(
    hovertemplate=
    "<b>%{label}</b><br>" +
    "Sales: %{value:.2f} million copies<br>" +
    "<extra></extra>"
)

fig.update_layout(
    width=1200,
    height=700
)

fig.show()
```

Let's dig deeper down within data by applying some filters

```python
# Let's find the top 10 best selling RPG games (global sales)
rpg_filter = vg_sales['Genre'].str.contains('role', case=False)

rpg_sales_global = vg_sales[rpg_filter]
rpg_sales_global[["Name", "Platform", "Genre", "Year", "Global_Sales"]].sort_values(by="Global_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling RPG games on PS2 (global sales)
rpg_filter = vg_sales['Genre'].str.contains('role', case=False)
platform_filter = vg_sales['Platform'].str.contains('ps2', case=False)

ps2_rpg_sales_global = vg_sales[rpg_filter & platform_filter]
ps2_rpg_sales_global[["Name", "Platform", "Genre", "Year", "Global_Sales"]].sort_values(by="Global_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling games for GameBoy Devices in North America
gameboy_filter = vg_sales['Platform'].str.contains('gb', case=False) # Please note: this way, we will find both GameBoy and GameBoy Advance sales data

gameboy_sales_na = vg_sales[gameboy_filter]
gameboy_sales_na[["Name", "Platform", "Genre", "Year", "NA_Sales"]].sort_values(by="NA_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling games for GameBoy (not GameBoy Advance etc.) in Japan
gameboy_excl_filter = vg_sales["Platform"].str.contains(r"^gb$", case=False, regex=True) # Using a RegEx, we can search for original GameBoy sales only 

gameboy_sales_jp = vg_sales[gameboy_excl_filter]
gameboy_sales_jp[["Name", "Platform", "Genre", "Year", "JP_Sales"]].sort_values(by="JP_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling games for GameBoy Advance in Europe
gameboy_adv__filter = vg_sales['Platform'].str.contains('gba', case=False)

gameboy_adv_sales_eu = vg_sales[gameboy_adv__filter]
gameboy_adv_sales_eu[["Name", "Platform", "Genre", "Year", "EU_Sales"]].sort_values(by="EU_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling Pokémon games in Japan
pokemon_filter = vg_sales['Name'].str.contains('pokemon', case=False)

pokemon_vg_sales_jp = vg_sales[pokemon_filter]
pokemon_vg_sales_jp[["Name", "Platform", "Genre", "Year", "JP_Sales"]].sort_values(by="JP_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling games in Japan, published by Capcom
capcom_filter = vg_sales['Publisher'].str.contains('capcom', case=False) # This filter wouldn't work if there still were missing values within the "Publisher" column

capcom_vg_sales_jp = vg_sales[capcom_filter]
capcom_vg_sales_jp[["Name", "Platform", "Publisher", "Genre", "Year", "JP_Sales"]].sort_values(by="JP_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best selling videogames in Europe, published by Sony Computer Entertainment
sony_filter = vg_sales['Publisher'].str.contains('sony', case=False) # This filter wouldn't work if there still were missing values within the "Publisher" column

sony_vg_sales_eu = vg_sales[sony_filter]
sony_vg_sales_eu[["Name", "Platform", "Publisher", "Genre", "Year", "EU_Sales"]].sort_values(by="EU_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best-selling videogames worldwide, published by Nintendo for NES (not SNES)
nintendo_filter = vg_sales['Publisher'].str.contains('nintendo', case=False) # This filter wouldn't work if there still were missing values within the "Publisher" column
nes_filter = vg_sales['Platform'].str.contains(r"^nes", case=False, regex=True) # Using a RegEx, we can search for NES sales only 

nintendo_nes_vg_sales_global = vg_sales[nintendo_filter & nes_filter]
nintendo_nes_vg_sales_global[["Name", "Platform", "Publisher", "Genre", "Year", "Global_Sales"]].sort_values(by="Global_Sales", ascending=False).round(1).head(10)
```

```python
# Let's find the top 10 best-selling videogames in Japan, published by Square for PlayStation consoles
square_filter = vg_sales['Publisher'].str.contains('square', case=False) # This filter wouldn't work if there still were missing values within the "Publisher" column
playstation_filter = vg_sales['Platform'].str.contains('ps', case=False)  

square_ps_vg_sales = vg_sales[square_filter & playstation_filter]
square_ps_vg_sales[["Name", "Platform", "Publisher", "Genre", "Year", "JP_Sales"]].sort_values(by="JP_Sales", ascending=False).round(1).head(10)
```
