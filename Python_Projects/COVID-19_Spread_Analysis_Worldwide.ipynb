```python
"""
COVID-19 spread analysis worldwide

The client requests a report on cases and vaccinations in various regions of the world; to this end, they ask that you use the dataset 
maintained by Our World in Data, available at https://github.com/owid/covid-19-data/tree/master/public/data; under the heading “Download our complete COVID-19 dataset," 
download the dataset in CSV or JSON format.

• Please verify the size of the dataset and the labels in the header.
• For each continent, please determine the total number of cases occurring on that continent; 
    please exclude any locations in the dataset that do not belong to any continent.
• Still regarding the total number of COVID cases, please develop a function that takes the dataset 
    and two continent names as input, and compares the following statistical descriptors: minimum and maximum values, 
    average, and percentage relative to the total number of cases worldwide (in this case, also calculated based on locations without a specified continent).
• You are then asked to perform the same type of analysis—again by developing a custom function—for the total number of vaccinations per continent.
• Finally, based on the calculations performed, the client asks you to write a brief (three or four lines) summary paragraph on the case and vaccination statistics, 
    focusing solely on the differences between Europe, South America, and Oceania.

To perform the analyses, we will use tools such as built-in functions and data structures like `open.()`; lists; and dictionaries;
built-in modules such as `cav` or `json`; and third-party modules such as Pandas and their methods and data types.
"""

import pandas as pd

covid_full_dataset = pd.read_csv("Python_Projects/CSV_Data_Repository/owid-covid-data.csv", sep=",")
covid_full_dataset
```

```python
# Let's check the dataset's dimensions (n. of rows and columns)
covid_full_dataset.shape

# Let's print column names
covid_full_dataset.columns
```

```python
# For each continent, determine the total number of cases that occurred on that continent; 
# Exclude any locations in the dataset that do not belong to any continent.

covid_continents = covid_full_dataset.loc[:,["continent", "total_cases", "total_vaccinations"]]
covid_filtered = covid_continents.groupby("continent")["total_cases"].max()
covid_filtered

# Alternative method:

covid_filtered = covid_full_dataset.loc[:,["continent", "total_cases", "total_vaccinations"]].groupby("continent")["total_cases"].max()
covid_filtered
```

```python
"""
With regard to total COVID cases, we are asked to develop a function that takes the dataset 
and the names of three continents as input, and compares the following statistical measures: minimum and maximum values, 
the average, and the percentage relative to the total number of cases worldwide (in this case, calculated to include locations without a specified continent).
"""

def statistical_function_cases():
    while True:
        cont_1 = str(input("Select a Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_1 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again!")
    while True:
        cont_2 = str(input("Select a second Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_2 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again!")
    while True:
        cont_3 = str(input("Select a third Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_3 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again!")

    data_cont_1 = covid_full_dataset["location"] == cont_1
    df_data_cont_1 = covid_full_dataset[data_cont_1]

    data_cont_2 = covid_full_dataset["location"] == cont_2
    df_data_cont_2 = covid_full_dataset[data_cont_2]

    data_cont_3 = covid_full_dataset["location"] == cont_3
    df_data_cont_3 = covid_full_dataset[data_cont_3]

    world_filter = covid_full_dataset["location"] == "World"
    df_data_cont_w = covid_full_dataset[world_filter]
    
    print("The minimum value of total cases in", cont_1,"is:", int(df_data_cont_1["total_cases"].min()))
    print("The minimum value of total cases in", cont_2,"is:", int(df_data_cont_2["total_cases"].min()))
    print("The minimum value of total cases in", cont_3 ,"is:", int(df_data_cont_3["total_cases"].min()),"\n")

    print("The maximum value of total cases in", cont_1,"is:", int(df_data_cont_1["total_cases"].max()))
    print("The maximum value of total cases in", cont_2,"is:", int(df_data_cont_2["total_cases"].max()))
    print("The maximum value of total cases in", cont_3,"is:", int(df_data_cont_3["total_cases"].max()),"\n")

    print("The average value of total cases in", cont_1,"is:", round(df_data_cont_1["total_cases"].mean(), 3))
    print("The average value of total cases in", cont_2,"is:", round(df_data_cont_2["total_cases"].mean(), 3))
    print("The average value of total cases in", cont_3,"is:", round(df_data_cont_3["total_cases"].mean(), 3),"\n")

    print("The total amount of cases in", cont_1 ,"accounts for ", round((df_data_cont_1["total_cases"].max() / df_data_cont_w["total_cases"].max()) * 100, 3), "% of cases worldwide.")
    print("The total amount of cases in", cont_2 ,"accounts for ", round((df_data_cont_2["total_cases"].max() / df_data_cont_w["total_cases"].max()) * 100, 3), "% of cases worldwide.")
    print("The total amount of cases in", cont_3 ,"accounts for ", round((df_data_cont_3["total_cases"].max() / df_data_cont_w["total_cases"].max()) * 100, 3), "% of cases worldwide.","\n")

statistical_function_cases()
```

```python
"""
Let's perform the same type of analysis—again by developing a custom function—for the total number of vaccinations on each continent.
"""

def statistical_function_vaccinations():
    while True:
        cont_1 = str(input("Select a Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_1 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again")
    while True:
        cont_2 = str(input("Select a second Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_2 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again")
    while True:
        cont_3 = str(input("Select a third Continent from the following: Africa, Asia, Europe, North America, Oceania, South America:\n"))
        if cont_3 in ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America']:
            break
        else: print("The Continent is not listed among the possible answers. Please try again")

    data_cont_1 = covid_full_dataset["location"] == cont_1
    df_data_cont_1 = covid_full_dataset[data_cont_1]

    data_cont_2 = covid_full_dataset["location"] == cont_2
    df_data_cont_2 = covid_full_dataset[data_cont_2]

    data_cont_3 = covid_full_dataset["location"] == cont_3
    df_data_cont_3 = covid_full_dataset[data_cont_3]

    world_filter = covid_full_dataset["location"] == "World"
    df_data_cont_w = covid_full_dataset[world_filter]

    print("The minimum value of vaccinations in", cont_1,"is:", int(df_data_cont_1["total_vaccinations"].min()))
    print("The minimum value of vaccinations in", cont_2,"is:", int(df_data_cont_2["total_vaccinations"].min()))
    print("The minimum value of vaccinations in", cont_3 ,"is:", int(df_data_cont_3["total_vaccinations"].min()),"\n")

    print("The maximum value of vaccinations in", cont_1,"is:", int(df_data_cont_1["total_vaccinations"].max()))
    print("The maximum value of vaccinations in", cont_2,"is:", int(df_data_cont_2["total_vaccinations"].max()))
    print("The maximum value of vaccinations in", cont_3,"is:", int(df_data_cont_3["total_vaccinations"].max()),"\n")

    print("The average value of vaccinations in", cont_1,"is:", round(df_data_cont_1["total_vaccinations"].mean(), 3))
    print("The average value of vaccinations in", cont_2,"is:", round(df_data_cont_2["total_vaccinations"].mean(), 3))
    print("The average value of vaccinations in", cont_3,"is:", round(df_data_cont_3["total_vaccinations"].mean(), 3),"\n")

    print("The total amount of vaccinations in", cont_1 ,"accounts for ", round((df_data_cont_1["total_vaccinations"].max() / df_data_cont_w["total_vaccinations"].max()) * 100, 3), "% of cases worldwide.")
    print("The total amount of vaccinations in", cont_2 ,"accounts for ", round((df_data_cont_2["total_vaccinations"].max() / df_data_cont_w["total_vaccinations"].max()) * 100, 3), "% of cases worldwide.")
    print("The total amount of vaccinations in", cont_3 ,"accounts for ", round((df_data_cont_3["total_vaccinations"].max() / df_data_cont_w["total_vaccinations"].max()) * 100, 3), "% of cases worldwide.","\n")
    
statistical_function_vaccinations()
```
