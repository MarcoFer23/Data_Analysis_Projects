```python
# Given the string: school_name = “Hogwarts”. Print each character of the string, one per line, using a for loop.

# Using a for loop:

school_name = "Hogwarts"
for x in school_name:
    print(x)

# using a while loop:

school_name = "Hogwarts"
i = 0
while i < len(school_name):
    print(school_name[i])
    i += 1
```

```python
"""
Given a list of chemical element symbols (all single letters) stored in the variable `elements = “NPKOHC”`,
print each element on a separate line.
"""

# Using a for loop:

elements = "NPKOHC"
for x in elements:
    print(x)

# Using a while loop:

elements = "NPKOHC"
i = 0
while i < len(elements):
    print(elements[i])
    i += 1

# Enriched versions:

# Using a for loop:

elements = "NPKOHC"
for x in elements:
    print("element symbol: ", x)

# Using a while loop:

elements = "NPKOHC"
i = 0
while i < len(elements):
    print("element symbol: ", elements[i])
    i += 1
```

```python
"""
Given a list of chemical element symbols (all single letters) stored in the variable `elements = “NPKOHC”`,
print each element on a separate line, preceded by the text “element - number n”, where n is a sequential number starting from 1.
"""

# Using a for loop:

elements = "NPKOHC"
for n, x in enumerate(elements,start=1):
    print("Element num. ",n,": ",x)

# Using a while loop:

elements = "NPKOHC"
i = 0
n = 1
while i < len(elements):
    print("Element num. ",n,": ",elements[i])
    i += 1
    n += 1
```

```python
"""
Modify the word “marmalade” so that all “e”s are replaced with “a”s and vice versa. Store the result in a variable and print it to the screen.

Create several versions:
• one using a for loop,
• one using a while loop,
• one using the string method .replace().
"""

# Using a for loop:

str_text = "marmalade"
new_str_text = ""

for x in str_text:
    if x == "a":
        new_str_text += "e"
    elif x == "e":
        new_str_text += "a"
    else:
        new_str_text += x 

print(new_str_text)               

# Using a while loop:

str_text = "marmalade"
i = 0
new_str_text = ""

while i < len(str_text):
    if str_text[i] == "a":
        new_str_text += "e"
    elif str_text[i] == "e":
        new_str_text += "a"
    else:
        new_str_text += str_text[i] 
    i += 1

print(new_str_text) 

# Using the .replace() string method

x = "marmalade"
num_a = x.count("a")

x = x.replace("e","a")
x = x.replace("a","e",num_a)

print(str(x))

# Alternative solution

str_text = "marmalade"
new_str_text = str_text.replace("e", "y").replace("a", "e").replace("y", "a")

print(new_str_text)
```

```python
"""
Calculate and print the first N powers of 2; each one should be stored at the end of a list.
Finally, print the resulting list.

Implement two versions:
• using a while loop,
• using a for loop.
"""

N = int(input("Enter an integer N."))

# Using a while loop:

x = 2
i = 0
new_list = list()

while i <= N:
    new_list.append(x**i)
    i += 1

print(new_list)

# Using a for loop:

x = 2

new_list = list()

# Using a for loop --> If we use “range,” the value N would not be included, so we write N + 1 to ensure that N itself is taken as the upper bound of the range:

for y in range(0,N+1): 
    new_list.append(x**y)    

print(new_list)
```

```python
"""
Calculate and print the first N powers of 3; each one should be added to the end of a list.
Finally, print the resulting list.

Implement two versions:
• using a while loop,
• using a for loop.
"""

N = int(input("Enter an integer N."))

# Using a while loop:

x = 3
i = 0
new_list = list()

while i <= N:
    new_list.append(x**i)
    i += 1

print(new_list)

# Using a for loop:

x = 3
new_list = list()

for y in range(0,N+1):
    new_list.append(x**y)    

print(new_list)
```
