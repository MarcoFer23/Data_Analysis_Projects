```python
"""
Let's import the `stocks` dataset from `plotly.express` using:

from plotly import express as px
stocks = px.data.stocks()

• Extract the data from the GOOG column regarding Google's stock performance,
and visualize it using pyplot

• Extract the first 5 rows from the GOOG column and the date column, and
use them as the x-axis and y-axis on a graph using pyplot

• Do the same for the last 5 rows of the dataset
"""

# Let's extract the data from the GOOG column showing Google's stock performance and plot it using pyplot
import matplotlib.pyplot as plt
from plotly import express as px

stocks = px.data.stocks()

google = stocks["GOOG"]

plt.plot(google, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.xlabel("Trend over time (days)")
plt.ylabel("Stock value")
plt.grid(True)
plt.show()

# Let's extract the first 5 rows from the “GOOG” column and the “date” column, and use them as the x-axis and y-axis on a graph using pyplot
y = stocks["GOOG"].head()
x = stocks["date"].head()

plt.plot(x, y, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.show()

# Let's do the same for the last 5 rows of the dataset
import matplotlib.pyplot as plt
from plotly import express as px

stocks = px.data.stocks()
y =stocks["GOOG"].tail()
x =stocks["date"].tail()

plt.plot(x, y, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.show()
```

```python
"""
Let's extract the first 20 rows from the AAPL column of Apple stock data and
plot the graph using pyplot so that:
• the graph is red
• the line is dashed
• there is a dot as the marker
• the x-axis is labeled “Date”
• the y-axis is labeled “Value”
• the graph title is “Apple Stock”
• the markerfacecolor is black
• the line thickness is 2
"""

stocks.head(20).plot(x = "date", y = "AAPL", color="r", linewidth=2, linestyle="--", marker="o", markersize=8.5, markerfacecolor="k", figsize=(10,6))
plt.title("Apple stocks")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(False)
plt.show()
```

```python
"""
We import the 'elections' dataset from plotly.express using:
from plotly import express as px
election = px.data.election()

• Create a bar chart comparing the total votes received by the three
candidates
• Display a comparative bar chart comparing the votes received
in the top 4 districts
• Display it in both paired and stacked formats
• Save both charts to disk in high resolution
"""

from plotly import express as px
election = px.data.election()

elections = px.data.election()
elections_set = elections.loc[:,["Coderre", "Bergeron", "Joly"]]
elections_set.sum().plot(kind="barh")
plt.savefig("test_1.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()

elections_set = elections.loc[0:3,["district", "Coderre", "Bergeron", "Joly"]]
elections_set = elections_set.set_index("district")
elections_set.plot(kind="barh")
plt.savefig("test_2.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()

elections_set.plot(kind="barh", stacked=True)
plt.legend(loc="lower right")
plt.savefig("test_3.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()
```

```python
"""
We import the Titanic dataset from Seaborn using:
import seaborn as sns
titanic = sns.load_dataset('titanic')

• How many decks were there on the ship?
• Is there any missing data? Where? How much? How could we handle it?
• Using Seaborn, we visualize a joint plot of the fare and age columns;
what are we looking at?
"""

import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic

set(titanic["deck"].dropna())

len(set(titanic["deck"].dropna()))

titanic.isnull().sum()
titanic.isna().sum()

"""
The column contains the “Categories” data type, so we can’t directly replace the missing data;
we need to add a new category to replace any Null and/or NaN values.
"""

titanic["deck"] = titanic["deck"].cat.add_categories("n.d.")
replaced = titanic["deck"].fillna("n.d.")
replaced 

sns.jointplot(data=titanic, x="age", y="fare", kind="hex")
```

```python
"""
• Create a graph showing the difference in the number of passengers for each
boarding class
• Do the same for the “alive” column
• What was the distribution of fares (calculate)?
• Can we see the age distribution of passengers relative to
boarding class? Let’s try a boxplot and a swarmplot
• Display a kdeplot for the age column
• Display a boxplot and an lmplot for the fare and
survived columns; what can we infer from this?
"""

titanic["class"].value_counts().sort_index().plot(kind="bar")
plt.show()

titanic["alive"].value_counts().sort_index().plot(kind="bar")
plt.show()

sns.displot(titanic.fare)
plt.show()

sns.kdeplot(titanic.fare)
plt.show()

sns.boxplot(data=titanic, x = "class", y = "age")
plt.show()

sns.swarmplot(data=titanic, x = "class", y = "age")
plt.show()

sns.kdeplot(titanic.age)
plt.show()

sns.boxplot(data=titanic, x = "survived", y = "fare")
plt.show()

sns.lmplot(data=titanic, x = "survived", y = "fare")
plt.show()
```

```python
"""
Let's create an interactive scatter plot using plotly.express where
we will have:
• age on the x-axis and deck on the y-axis
• color based on whether the person survived or not
• marker size based on fare
• the symbol (using the symbol= parameter) based on the sex column
• we reorder the y-axis to match the arrangement of the decks, using the
parameter category_orders={“deck”: list(“ABCDEFG”)}
"""

from plotly import express as px
px.scatter(titanic, x = "age" , y = "deck", color="survived", size="fare", symbol="sex", category_orders={"deck": list("ABCDEFG")})
```
