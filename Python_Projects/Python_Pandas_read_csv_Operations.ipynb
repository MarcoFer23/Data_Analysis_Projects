```python
"""
Let’s download the Iris dataset: https://archive.ics.uci.edu/dataset/53/iris
We will find a .data file, which is a CSV, and a .names file containing metadata; this
version of the dataset does not have column names.
• Read the file and load it into a DataFrame using pd.read_csv()
without using any other parameters
• Print the first five rows
• Print the column names: are they correct?
7.2
From the previous dataset:
• Read the file and load it into a DataFrame, adding the
column names — found in the .names file — as a parameter to
pd.read_csv()
• Print the first five rows and the last ten
• Print a summary of the dataset’s descriptive statistics
"""

import pandas as pd
data_set = pd.read_csv("iris.data")

# Let's print the first 5 rows
print(data_set.head())

# Let's print column names
print(data_set.columns) # Column names aren't correct

# Let's read the file and load it into a DataFrame, adding the column names—which are found in the .names file—as a parameter to pd.read_csv().
data_set = pd.read_csv("iris.data",header=None,names=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm','class'])

# Let's print the first 5 rows
data_set.head() # Per default vengono stampate le prime cinque righe utilizzando .head()

# Let's print the last 10 rows
data_set.tail(10)

# Let's print a summary of the descriptive statistics for the dataset
data_set.describe()
```
