```python
# Reading amazon database downloaded from Kaggle

import pandas as pd

db = pd.read_csv("File_Data_Repository/amazon.csv", sep=",")
amazon_db = pd.DataFrame(db)
amazon_db
```

```python
# Let's print the database's shape, then let's see a sample of 10 rows and the column names

print(amazon_db.shape, "\n")
print(amazon_db.sample(10), "\n")
print(amazon_db.columns, "\n")
```

```python
# Calculate negative reviews
print(amazon_db['Positive'].value_counts(),"\n")

# Alternative method(s)
tot_reviews = amazon_db['Positive'].value_counts()
print("Positive reviews:", tot_reviews.get(1, 0))
print("Negative reviews:", tot_reviews.get(0, 0))
print() # To add a new line

print("Please note: the column 'Positive' contains boolean values, so that 1 = True and 0 = False")

# Conditional check
pos_reviews = tot_reviews.get(1, 0)
neg_reviews = tot_reviews.get(0, 0)

# Sentiment evaluation
if pos_reviews > neg_reviews:
    print("\nThe amount of positive reviews is higher:", pos_reviews," vs ", neg_reviews)
elif neg_reviews > pos_reviews:
    print("\nThe amount of negative reviews is higher:", neg_reviews," vs ",pos_reviews)
else:
    print("\nThe amount of positive and negative reviews is equal.")
```

```python
# Search for reviews containing the word "app"
filter_app = amazon_db['reviewText'].str.contains('app', case=False)

reviews_cont_app = amazon_db[filter_app]
reviews_cont_app
```

```python
# Search for reviews containing the word "game"
filter_game = amazon_db['reviewText'].str.contains('game', case=False)

reviews_cont_game = amazon_db[filter_game]
reviews_cont_game
```

```python
# Count values in the "Positive" column
value_counts = amazon_db['Positive'].value_counts()
value_counts
```

```python
# Search for positive reviews containing the word "recommend"; print a sample of 10 rows and order by the index
filter_recommend = amazon_db['reviewText'].str.contains('recommend', case=False)
filter_positive = amazon_db['Positive'] == True

recommend_positive_reviews = amazon_db[filter_recommend & filter_positive]
recommend_positive_reviews.sample(10).sort_index()
```

```python
# Search for negative reviews; print a sample of 10 rows and order by the index
filter_negative = amazon_db['Positive'] == False

negative_reviews = amazon_db[filter_negative]
negative_reviews.sample(10).sort_index()
```
