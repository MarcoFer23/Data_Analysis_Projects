```python
"""
Quesito 1 - Parte 1
Da plotly.express importiamo il dataset stocks, mediante:
from plotly import express as px
stocks = px.data.stocks()

• Estraiamo i dati della colonna GOOG relative all'andamento delle azioni di
Google, e visualizziamolo mediante pyplot

• Estraiamo le prime 5 righe della colonna GOOG e della colonna date, e
usiamoli come ascisse e ordinate su un grafico mediante pyplot

• Facciamo lo stesso per le ultime 5 righe del dataset
"""

# Estraiamo i dati della colonna GOOG relative all'andamento delle azioni di Google, e visualizziamolo mediante pyplot
import matplotlib.pyplot as plt
from plotly import express as px

stocks = px.data.stocks()

google = stocks["GOOG"]

plt.plot(google, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.xlabel("Andamento nel tempo (giorni)")
plt.ylabel("Valore azioni")
plt.grid(True)
plt.show()

# Estraiamo le prime 5 righe della colonna GOOG e della colonna date, e usiamoli come ascisse e ordinate su un grafico mediante pyplot
y = stocks["GOOG"].head()
x = stocks["date"].head()

plt.plot(x, y, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.show()

# Facciamo lo stesso per le ultime 5 righe del dataset
import matplotlib.pyplot as plt
from plotly import express as px

stocks = px.data.stocks()
y =stocks["GOOG"].tail()
x =stocks["date"].tail()

plt.plot(x, y, "grey", linewidth=1.5, marker=".", markersize=8.5, markerfacecolor="black")
plt.title("Google Stocks: GOOG")
plt.show()
```

```python
"""
Quesito 1 - Parte 2
Estraiamo le prime 20 istanze della colonna AAPL delle azioni di Apple, e
visualizziamo il grafico tramite pyplot, in modo che:
• il grafico sia rosso
• la linea sia tratteggiata
• vi sia un pallino come marker
• l'asse delle ascisse si chiami "Data"
• l'asse delle ordinate si chiami "Valore"
• il titolo del grafico sia "Azioni Apple"
• il markerfacecolor sia nero
• la linea abbia spessore uguale a 2
"""

stocks.head(20).plot(x = "date", y = "AAPL", color="r", linewidth=2, linestyle="--", marker="o", markersize=8.5, markerfacecolor="k", figsize=(10,6))
plt.title("Azioni Apple")
plt.xlabel("Data")
plt.ylabel("Valore")
plt.grid(False)
plt.show()
```

```python
"""
Quesito 2
Da plotly.express importiamo il dataset elections, mediante:
from plotly import express as px
election = px.data.election()

• Disegnamo un grafico a barre dove confrontiamo i voti totali presi dai tre
candidati
• Visualizzare un grafico a barre comparativo dove si confrontano i voti presi
nei primi 4 distretti
• Visualizzarlo sia in formato appaiato che impilato (stacked)
• Salvare entrambi i grafici su disco in alta risoluzione
"""

elections = px.data.election()
elections_set = elections.loc[:,["Coderre", "Bergeron", "Joly"]]
elections_set.sum().plot(kind="barh")
plt.savefig("prova_1.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()

elections_set = elections.loc[0:3,["district", "Coderre", "Bergeron", "Joly"]]
elections_set = elections_set.set_index("district")
elections_set.plot(kind="barh")
plt.savefig("prova_2.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()

elections_set.plot(kind="barh", stacked=True)
plt.legend(loc="lower right")
plt.savefig("prova_3.png", dpi=600, bbox_inches = 'tight', pad_inches=0.5) 
plt.show()
```

```python
"""
Quesito 3 - Parte 1
Da seaborn importiamo il dataset titanic, mediante:
import seaborn as sns
titanic = sns.load_dataset('titanic')

• Quanti ponti c'erano sulla nave?
• Ci sono dati mancanti? Dove? Quanti? Come potremmo gestirli?
• Tramite seaborn visualizziamo un jointplot sulle colonne fare e age;
che cosa stiamo guardando?
"""

import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic

set(titanic["deck"].dropna())

len(set(titanic["deck"].dropna()))

titanic.isnull().sum()
titanic.isna().sum()

titanic["deck"] = titanic["deck"].cat.add_categories("n.d.")
replaced = titanic["deck"].fillna("n.d.")
replaced 

"""
La colonna contiene datatype Categories quindi non si può sostituire direttamente il dato mancante, 
va aggiunta una nuova categoria per sostituire eventuali Null e/o NaN.
"""

sns.jointplot(data=titanic, x="age", y="fare", kind="hex")
```

```python
"""
Quesito 3 - Parte 2
• Visualizzare un grafico con la differenza tra il numero di passeggeri di ogni
classe di imbarco
• Fare la stessa cosa per la colonna alive
• Qual era la distribuzione delle tariffe (fare)?
• Riusciamo a vedere la distribuzione delle età dei passeggeri rispetto alla
classe di imbarco? Proviamo con un boxplot e con uno swarmplot
• Visualizziamo un kdeplot per la colonna age
• Visualizziamo un boxplot e un lmplot rispetto alle colonne fare e
survived; che cose ne deduciamo?
"""

titanic["class"].value_counts().sort_index().plot(kind="bar")
plt.show()

titanic["alive"].value_counts().sort_index().plot(kind="bar")
plt.show()

sns.distplot(titanic.fare)
plt.show()

sns.kdeplot(titanic.fare)
plt.show()

sns.boxplot(data=titanic, x = "class", y = "age")
plt.show()

sns.swarmplot(data=titanic, x = "class", y = "age")
plt.show()

sns.kdeplot(titanic.age)
plt.show()

sns.boxplot(data=titanic, x = "survived", y = "fare")
plt.show()

sns.lmplot(data=titanic, x = "survived", y = "fare")
plt.show()
```

```python
"""
Quesito 3 - Parte 3
Realizziamo un grafico scatter interattivo tramite plotly.express dove
avremo:
• l'età sull'asse x e il ponte sull'asse y
• il colore rispetto a se la persona è sopravvissuta o meno
• la dimensione del marker rispetto alla tariffa
• il simbolo (tramite il parametro symbol=) rispetto alla colonna sex
• riordiniamo l'asse y in modo che rispetti la disposizione dei ponti, mediante il
parametro category_orders={"deck": list("ABCDEFG")}

Che conclusioni possiamo trarne?
"""

from plotly import express as px
px.scatter(titanic, x = "age" , y = "deck", color="survived", size="fare", symbol="sex", category_orders={"deck": list("ABCDEFG")})
```
