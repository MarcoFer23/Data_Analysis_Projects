```python
"""
Esercizio 1 - Parte 2
"""

import pandas as pd

db = pd.read_csv("C:/Users/marco/Downloads/archive/beginner_datasets/amazon.csv", sep=",")
amazon_db = pd.DataFrame(db)
amazon_db

colonna = 'reviewText'
ricerca_valore = 'app'
filtro = amazon_db[colonna].str.contains(ricerca_valore, case=False)
risultato = amazon_db[filtro].sum()
risultato

colonna = 'reviewText'
ricerca_valore = 'game'
filtro = amazon_db[colonna].str.contains(ricerca_valore, case=False)
risultato = amazon_db[filtro].sum()
risultato

colonna = 'Positive'  
conteggio_valori = amazon_db[colonna].value_counts()
conteggio_valori

filtro_recommend = amazon_db['reviewText'].str.contains('recommend', case=False)
filtro_positive = amazon_db['Positive'] == True
risultato = amazon_db[ filtro_recommend & filtro_positive ].value_counts()
risultato
risultato.sample(10)

filtro_positive = amazon_db['Positive'] == False
risultato = amazon_db[filtro_positive].value_counts()
risultato
risultato.sample(10)

filtro_all = amazon_db['reviewText'].str.contains('kindle', case=False)
risultato_overall = amazon_db[filtro_all]
print(risultato_overall.shape[0])

filtro = amazon_db['reviewText'].str.contains('kindle', case=False)
filtro_positive = amazon_db['Positive'] == True
risultato_positive = amazon_db[filtro & filtro_positive]
print(risultato_positive.shape[0])

risultato_negative = risultato_overall.shape[0] - risultato_positive.shape[0]
print(risultato_negative)

if risultato_negative < risultato_positive.shape[0]:
    print("Le recensioni sono perlopiù positive.")
else:
    print("Le recensioni sono perlopiù negative.")
```
