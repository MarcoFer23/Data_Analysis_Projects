```python
"""
Esercizio 3
Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini
delle persone, e della zona in cui vivono, rispetto ai costi individuali per le cure
mediche come premio per le assicurazioni sulla salute.

Visualizziamone le dimensioni, un'anteprima e osserviamo i nomi di colonna;
Quali sono le medie di charges rispetto a region? Ci sono differenze
significative?
E rispetto a smoker? E a sex?
Quali sono i descrittori statistici di bmi? 
Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
"""

import pandas as pd
import numpy as np

db = pd.read_csv("C:/Users/marco/Downloads/archive/beginner_datasets/insurance.csv", sep=",")
insurance = pd.DataFrame(db)


print("Il dataset \'insurance.csv\' contiene i seguenti dati:\n", insurance)
print("\nIl dataset ha le seguenti dimensioni (numero rispettivamente di righe e colonne):\n", insurance.shape)
print("\nDi seguito si riporta un'anteprima randomica dei dati contenuti nel dataset:\n", insurance.sample(10))
print("\nIl dataset è composto dalle seguenti colonne (di cui si riportano i nomi):\n",insurance.columns)

# 1. Medie di charges rispetto a region
media_charges_region = insurance.groupby('region')['charges'].mean()
print("\nDi seguito si riportano le medie di \'charges\' rispetto a \'region\':\n")
print(media_charges_region)

# 2. Differenze significative rispetto a smoker
media_charges_smoker = insurance.groupby('smoker')['charges'].mean()
print("\nDi seguito si riportano le medie di \'charges\' rispetto a \'smoker\':\n")
print(media_charges_smoker)

# 3. Differenze significative rispetto a sex
media_charges_sex = insurance.groupby('sex')['charges'].mean()
print("\nDi seguito si riportano le medie di \'charges\' rispetto a \'sex\':\n")
print(media_charges_sex)

# 4. Descrittori statistici di bmi
bmi_stats = insurance['bmi'].describe()
print("\nDi seguito si riportano i descrittori statistici rispetto a \'bmi\':\n")
print(bmi_stats)

# 5. Minimo, media e massimo di charges rispetto ai diversi quartili di bmi
quartili = np.percentile(insurance['bmi'], [25, 50, 75])
diff_quartili = insurance.groupby(pd.cut(insurance['bmi'], bins=[0, quartili[0], quartili[1], quartili[2], float('inf')]))['charges'].agg(['min', 'mean', 'max'])
print("\nDi seguito si riportano Minimo, Media e Massimo di \'charges\' rispetto ai diversi quartili di \'bmi\':")
print(diff_quartili)
```

```python
"""
Esercizio 4 - Parte 2
• Ci sono ripetizioni, ovvero righe che si riferiscono allo stesso Pokémon?
• Per ogni tipo in Type 1 quanti ce ne sono?
• Quali sono i Pokémon leggendari?
• E quali sono i leggendari di tipo 1 Grass?
• E leggendari di tipo 1 Ice o Fire?
• Ordiniamo il dataset per la colonna Name
• Trasformiamo Name nell'indice
• Come accediamo alle statistiche di tutti i Pokémon il cui nome inizia per D?
• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
• Ci sono dati nulli?
"""

import pandas as pd

db = pd.read_csv("C:/Users/marco/Downloads/archive/beginner_datasets/pokemon.csv", sep=",")
pokemon = pd.DataFrame(db)
pokemon

# Ci sono ripetizioni, ovvero righe che si riferiscono allo stesso Pokémon?
duplicati = pokemon["#"].duplicated()
if duplicati.empty:
    print("Non ci sono ripetizioni nel dataset.")
else:
    print("Ci sono ripetizioni nel dataset.")

# Per ogni tipo in Type 1 quanti ce ne sono?
contegg_pokemon_type1 = pokemon["Type 1"].value_counts()
print("\nConteggio dei Pokémon \'Type 1\':\n")
print(contegg_pokemon_type1)

# Quali sono i Pokémon leggendari?
pokemon_legg = pokemon[pokemon['Legendary'] == True]
print("\nDi seguito si riportano i nomi di tutti i Pokémon Leggendari:\n")
print(pokemon_legg[['Name', 'Type 1', 'Legendary']])

# E quali sono i leggendari di tipo 1 Grass?
legendary_grass = pokemon_legg[pokemon_legg['Type 1'] == 'Grass']
print("\nDi seguito di riportano i nomi di tutti i Pokémon Leggendari di tipo \'Erba\':\n")
print(legendary_grass[['Name', 'Type 1', 'Legendary']])

# E leggendari di tipo 1 Ice o Fire?
legendary_ice_fire = pokemon_legg[(pokemon_legg['Type 1'] == 'Ice') | (pokemon_legg['Type 1'] == 'Fire')]
print("\nDi seguito di risportano i nomi di tutti i Pokémon Leggendari di tipo \'Ghiaccio\' o \'Fuoco\':\n")
print(legendary_ice_fire[['Name', 'Type 1', 'Legendary']])

# Ordiniamo il dataset per la colonna Name
pokemon.sort_values(by='Name', inplace=True)

# Trasformiamo Name nell'indice
pokemon.set_index('Name', inplace=True)

# Come accediamo alle statistiche di tutti i Pokémon il cui nome inizia per D?
nomi_pokemon_iniziale_d = pokemon[pokemon["Name"].str.startswith('D')]
print("\nDi seguito si riportano i nomi dei Pokémon il cui nome inizia per 'D':\n")
print(nomi_pokemon_iniziale_d['Name'])

# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
filtro_generazione = pokemon[pokemon['Generation'] == 1]
filtro_pokemon = filtro_generazione[(filtro_generazione['Attack'] > 50) & (filtro_generazione['HP'] < 60)]
print("\nDi seguito si riportano i nomi dei Pokémon di prima generazione con attacco > 50 e HP < 60:\n")
print(filtro_pokemon['Name'])

# Ci sono dati nulli?
null_data = pokemon.isnull().sum()
print("\nDi seguito si riporta il conteggio dei dati nulli per ogni colonna all'interno del dataset:\n")
print(null_data)
```
